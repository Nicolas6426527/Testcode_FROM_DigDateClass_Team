```python
from selenium import webdriver
from time import sleep

import json

import os,re
```

```python
linkdata1 = []
name =[]
link={}

driver = webdriver.Chrome("/Users/Apple/pyds/chromedriver")
driver.get("https://www.gutenberg.org/browse/languages/zh?fbclid=IwAR1VOR2GxDBwRKQkIgWRbP6MDeC6wdekM4yteUTCIIgE6j-dOLTm4m_Jrdg")
link1 = driver.find_elements_by_css_selector("li.pgdbetext a")

for a in link1:
    link2 = a.get_attribute("href")
    b=a.text
    linkdata1.append(link2)
    name.append(b)
    
```

```python
title={}
for i in range(len(name)) :
    title[name[i]] = linkdata1[i] #500==500
print(title)
```

```python
def reFilename(text):
    comp = re.compile('[，。：「」；、？！『』:=!,\n]+')#把想要換的符號給marker起來，在使用sub把marker的字符換掉
    return comp.sub('',text).replace('.','_')#把上面的符號換成下面的“”＝是沒有空格的意思 ,text = 加入上述的字串
#在把.的部分轉換成＿避免window在讀取時把.讀成檔案
```

```python

for  key in title :
    driver.get(title[key])
    b = driver.find_element_by_css_selector("a[type=\"text/plain; charset=utf-8\"]")
    if len(b)>0 :#清單，沒抓到是0
        driver.get(b[0].get_attribute("href"))#回傳是一個清單，而只有一個
        filename=reFilename(key)
        with  open(f'./novel/{filename}.txt',"w",encoding = "utf-8") as f:
             f.write(driver.find_element_by_css_selector("pre[style='word-wrap: break-word; white-space: pre-wrap;']").text)
    
```
